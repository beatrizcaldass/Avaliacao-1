# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é proporcional a área sob a curva de crescimento do cacto ao longo do tempo. Determine a área sob a curva da função $y = x^5(2 - x)$ nestes dois anos.



##**RESPOSTA 1**

Configurei a integral:
$$ \text{Área} = \int_{0}^{2} x^5(2 - x) \, dx $$

Dividi a integral em duas partes:
$$ \text{Área} = \int_{0}^{2} 2x^5 \, dx - \int_{0}^{2} x^6 \, dx $$

Calculei cada parte da integral:
$$ \int 2x^5 \, dx = 2 \int x^5 \, dx = 2 \cdot \frac{x^6}{6} = \frac{1}{3} x^6 $$
$$ \int x^6 \, dx = \frac{x^7}{7} $$

Avaliei as integrais nos limites de \( 0 \) a \( 2 \):
$$ \left[ \frac{1}{3} x^6 \right]_{0}^{2} = \frac{1}{3} (2^6) - \frac{1}{3} (0^6) = \frac{64}{3} $$
$$ \left[ \frac{x^7}{7} \right]_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7} = \frac{128}{7} $$

Subtrai os resultados das integrais:
$$ \text{Área} = \frac{64}{3} - \frac{128}{7} $$

Encontrei um denominador comum e realizei a subtração:
$$ \text{Área} = \frac{64 \cdot 7}{21} - \frac{128 \cdot 3}{21} $$
$$ \text{Área} = \frac{448}{21} - \frac{384}{21} $$
$$ \text{Área} = \frac{64}{21} $$

Portanto, a área sob a curva da função $( y = x^5(2 - x) )$ nos primeiros 2 anos é $( \frac{64}{21} \approx 3.0476 )$ unidades de área.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')

f = x**5 * (2 - x)

area = sp.integrate(f, (x, 0, 2))

f_np = sp.lambdify(x, f, 'numpy')

x_vals = np.linspace(0, 2, 400)
y_vals = f_np(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$y = x^5(2 - x)$')
plt.fill_between(x_vals, y_vals, where=[(xi >= 0 and xi <= 2) for xi in x_vals], color='skyblue', alpha=0.4, label=f'Área = {area.evalf():.4f}')
plt.title('Área sob a curva $y = x^5(2 - x)$ no intervalo [0, 2]')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.show()

print(f'A área sob a curva é: {area.evalf():.4f}')
```

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F9x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

##**RESPOSTA 2**

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, diff, solve

def F(x):
    return x**5 * (2 - x)

x = np.linspace(0, 2, 400)

y = F(x)

x_sym = symbols('x')
F_sym = x_sym**5 * (2 - x_sym)
area = integrate(F_sym, (x_sym, 0, 2))

derivative = diff(F_sym, x_sym)
critical_points = solve(derivative, x_sym)
critical_points = [point.evalf() for point in critical_points if 0 <= point.evalf() <= 2]

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$F(x) = x^5 (2 - x)$')
plt.fill_between(x, y, where=[(xi >= 0 and xi <= 2) for xi in x], color='skyblue', alpha=0.4, hatch='//')
plt.title('Gráfico de $F(x) = x^5 (2 - x)$ e a Área sob a Curva')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.legend()
plt.grid(True)

for point in critical_points:
    if point.is_real:
        y_value = F(point).evalf()
        plt.annotate(f'({point:.2f}, {y_value:.2f})', xy=(float(point), float(y_value)), xytext=(float(point), float(y_value) + 0.5),
                     arrowprops=dict(facecolor='black', arrowstyle='->', shrinkA=0, shrinkB=5))

plt.annotate(f'Área = {area.evalf():.4f}', xy=(1, 0.5), xytext=(1.5, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->', shrinkA=0, shrinkB=5))


plt.show()
print(f'A área sob a curva é: {area.evalf():.4f}')
```

# Questão 3 - Vale até 1 ponto

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele percorreu.

##**RESPOSTA 3**



Função da velocidade $(v(t))$:
$$ a(t) = \frac{dv}{dt} $$
Precisa integrar a aceleração para obter a velocidade:
$$ v(t) = \int \cos(\pi t) \, dt $$
A integral de \(\cos(\pi t)\) é:
$$ v(t) = \frac{1}{\pi} \sin(\pi t) + C $$

Usando a condição inicial $(v(0) = \frac{1}{2\pi})$, encontrei a constante (C):
$$ v(0) = \frac{1}{\pi} \sin(0) + C = \frac{1}{2\pi} $$
$$ C = \frac{1}{2\pi} $$
Portanto, a função da velocidade é:
$$ v(t) = \frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi} $$

Função da posição \(s(t)\):
$$ v(t) = \frac{ds}{dt} $$
Portanto, precisa integrar a velocidade para obter a posição:
$$ s(t) = \int v(t) \, dt = \int \left(\frac{1}{\pi} \sin(\pi t) + \frac{1}{2\pi}\right) dt $$
A integral de $(\frac{1}{\pi} \sin(\pi t))$ é:
$$ \int \frac{1}{\pi} \sin(\pi t) \, dt = -\frac{1}{\pi^2} \cos(\pi t) $$
E a integral de $(\frac{1}{2\pi})$ é:
$$ \int \frac{1}{2\pi} \, dt = \frac{1}{2\pi} t $$
Portanto, a função da posição é:
$$ s(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + D $$
Se usa a condição inicial \(s(0) = 0\) para encontrar a constante \(D\):
$$ s(0) = -\frac{1}{\pi^2} \cos(0) + \frac{1}{2\pi} \cdot 0 + D = 0 $$
$$ D = \frac{1}{\pi^2} $$
Portanto, a função da posição é:
$$ s(t) = -\frac{1}{\pi^2} \cos(\pi t) + \frac{1}{2\pi} t + \frac{1}{\pi^2} $$

Calculei a posição em \(t = 1.5\):
$$ s(1.5) = -\frac{1}{\pi^2} \cos(\pi \cdot 1.5) + \frac{1}{2\pi} \cdot 1.5 + \frac{1}{\pi^2} $$
Sabemos que $(\cos(\pi \cdot 1.5) = \cos(1.5\pi) = 0)$, portanto:
$$ s(1.5) = -\frac{1}{\pi^2} \cdot 0 + \frac{1.5}{2\pi} + \frac{1}{\pi^2} $$
$$ s(1.5) = \frac{0.75}{\pi} + \frac{1}{\pi^2} $$

 A distância total percorrida pelo objeto nos primeiros 1,5 segundos é:
$$ s(1.5) - s(0) = \left( \frac{0.75}{\pi} + \frac{1}{\pi^2} \right) - \frac{1}{\pi^2} = \frac{0.75}{\pi} $$
Portanto, a distância total percorrida é:
$$ \frac{0.75}{\pi} \approx 0.2387 \, \text{unidades de comprimento} $$


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

t = sp.symbols('t')

a_t = sp.cos(sp.pi * t)
v_t = sp.integrate(a_t, t) + 1/(2*sp.pi)

a_t_np = sp.lambdify(t, a_t, 'numpy')
v_t_np = sp.lambdify(t, v_t, 'numpy')

t_vals = np.linspace(0, 1.5, 400)
a_vals = a_t_np(t_vals)
v_vals = v_t_np(t_vals)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(t_vals, a_vals, label='$a(t) = \cos(\pi t)$')
plt.title('Aceleração $a(t)$')
plt.xlabel('Tempo $t$ (s)')
plt.ylabel('Aceleração $a(t)$')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(t_vals, v_vals, label='$v(t)$')
plt.title('Velocidade $v(t)$')
plt.xlabel('Tempo $t$ (s)')
plt.ylabel('Velocidade $v(t)$')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de bibliotecas definidas para esta avaliação.  


##**RESPOSTA 4**

```python
import sympy as sp

x = sp.symbols('x')

f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3

h = f - g

area = sp.integrate(h, (x, 1, 2))
area_eval = area.evalf()

print(f'A área limitada pelas funções f(x) e g(x) no intervalo [1, 2] é: {area_eval:.4f} unidades quadradas')
```

Defini as funções:
$$ f(x) = -x^2 + 4x + 1 $$
$$ g(x) = -x^3 + 7x^2 - 10x + 3 $$

Calculei a diferença \( h(x) = f(x) - g(x) \):
$$ h(x) = (-x^2 + 4x + 1) - (-x^3 + 7x^2 - 10x + 3) $$
$$ h(x) = -x^2 + 4x + 1 + x^3 - 7x^2 + 10x - 3 $$
$$ h(x) = x^3 - 8x^2 + 14x - 2 $$

A área sob a curva de \( h(x) \) no intervalo \( [1, 2] \) é dada por:
$$ \text{Área} = \int_1^2 h(x) \, dx $$
$$ \text{Área} = \int_1^2 (x^3 - 8x^2 + 14x - 2) \, dx $$

Calculei a integral de cada termo:
$$ \int x^3 \, dx = \frac{x^4}{4} $$
$$ \int -8x^2 \, dx = -8 \cdot \frac{x^3}{3} = -\frac{8x^3}{3} $$
$$ \int 14x \, dx = 14 \cdot \frac{x^2}{2} = 7x^2 $$
$$ \int -2 \, dx = -2x $$

Somei as integrais, obtive:
$$ \int (x^3 - 8x^2 + 14x - 2) \, dx = \frac{x^4}{4} - \frac{8x^3}{3} + 7x^2 - 2x + C $$

Avaliaremos a integral nos limites de 1 a 2:
$$ \left[ \frac{x^4}{4} - \frac{8x^3}{3} + 7x^2 - 2x \right]_1^2 $$

Primeiro, avaliei em \( x = 2 \):
$$ \left. \frac{x^4}{4} - \frac{8x^3}{3} + 7x^2 - 2x \right|_{x=2} = \frac{2^4}{4} - \frac{8 \cdot 2^3}{3} + 7 \cdot 2^2 - 2 \cdot 2 $$
$$ = \frac{16}{4} - \frac{64}{3} + 28 - 4 $$
$$ = 4 - \frac{64}{3} + 28 - 4 $$
$$ = 28 - \frac{64}{3} $$

Depois, avaliei em \( x = 1 \):
$$ \left. \frac{x^4}{4} - \frac{8x^3}{3} + 7x^2 - 2x \right|_{x=1} = \frac{1^4}{4} - \frac{8 \cdot 1^3}{3} + 7 \cdot 1^2 - 2 \cdot 1 $$
$$ = \frac{1}{4} - \frac{8}{3} + 7 - 2 $$
$$ = 7 - 2 - \frac{8}{3} + \frac{1}{4} $$
$$ = 5 - \frac{8}{3} + \frac{1}{4} $$

Subtrai as duas avaliações para encontrar a área:
$$ \text{Área} = \left(28 - \frac{64}{3}\right) - \left(5 - \frac{8}{3} + \frac{1}{4}\right) $$
$$ = 28 - 5 - \frac{64}{3} + \frac{8}{3} - \frac{1}{4} $$
$$ = 23 - \frac{56}{3} - \frac{1}{4} $$
$$ = 23 - 18.67 - 0.25 $$
$$ = 23 - 18.92 $$
$$ = 4.08 $$

Portanto, a área sob a curva entre \( f(x) \) e \( g(x) \) no intervalo \( [1, 2] \) é aproximadamente \( 4.08 \) unidades de área.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

x = sp.symbols('x')

f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3

f_np = sp.lambdify(x, f, 'numpy')
g_np = sp.lambdify(x, g, 'numpy')

x_vals = np.linspace(1, 2, 400)
f_vals = f_np(x_vals)
g_vals = g_np(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='$f(x) = -x^2 + 4x + 1$')
plt.plot(x_vals, g_vals, label='$g(x) = -x^3 + 7x^2 - 10x + 3$')
plt.fill_between(x_vals, f_vals, g_vals, where=(f_vals > g_vals), interpolate=True, color='skyblue', alpha=0.4, label='Área entre f(x) e g(x)')

x_points = [1, 2]
for i in x_points:
    plt.plot(i, f_np(i), 'ro')
    plt.plot(i, g_np(i), 'bo')
    plt.annotate(f'({i}, {f_np(i):.2f})', xy=(i, f_np(i)), xytext=(i+0.1, f_np(i)+1),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)
    plt.annotate(f'({i}, {g_np(i):.2f})', xy=(i, g_np(i)), xytext=(i+0.1, g_np(i)-1),
                 arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10)

plt.title('Área entre as funções $f(x)$ e $g(x)$ no intervalo $[1, 2]$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend()
plt.grid(True)
plt.show()
```

# Questão 5 - Vale até 1 ponto

Use o método dos discos para determinar o volume do sólido obtido pela rotação da região delimitada por $y = \sqrt{x}$, $y = 3$ e o próprio eixo $y$ em torno do eixo $y$.


##**RESPOSTA 5**

Volume do sólido obtido pela rotação da região delimitada por $( y = \sqrt{x}),$ ( y = 3 ) e o eixo ( y )

Para encontrar o volume do sólido, tem que usar o método dos discos. A região é delimitada pelas curvas $$( y = \sqrt{x} )$$, ( y = 3) e o eixo ( y).

Expressei \( x \) em termos de \( y \):
   $$ x = y^2 $$

Configurei a integral do volume usando o método dos discos:
   $$ V = \pi \int_{0}^{3} (y^2)^2 \, dy $$

Calculei a integral:
   $$ V = \pi \int_{0}^{3} y^4 \, dy $$

A integral de $( y^4 )$ é:
   $$ \int y^4 \, dy = \frac{y^5}{5} $$

Analisei a integral nos limites de 0 a 3:
   $$ V = \pi \left[ \frac{y^5}{5} \right]_{0}^{3} $$
   $$ V = \pi \left( \frac{3^5}{5} - \frac{0^5}{5} \right) $$
   $$ V = \pi \left( \frac{243}{5} \right) $$
   $$ V = \frac{243\pi}{5} $$

Portanto, o volume do sólido obtido pela rotação da região delimitada por $( y = \sqrt{x} )$, $( y = 3 )$ e o eixo $( y )$ em torno do eixo $( y )$ é $(\frac{243\pi}{5})$ unidades cúbicas, aproximadamente 152.6814.


```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

y = sp.symbols('y')

x = y**2

y_min = 0
y_max = 3

volume = sp.integrate(sp.pi * x**2, (y, y_min, y_max))

volume_val = volume.evalf()

y_vals = np.linspace(y_min, y_max, 400)
x_vals = np.sqrt(y_vals)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

ax[0].fill_betweenx(y_vals, x_vals, color='skyblue', alpha=0.4)
ax[0].plot(x_vals, y_vals, label='$y = \sqrt{x}$')
ax[0].hlines(y=3, xmin=0, xmax=3, color='r', label='$y = 3$')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$')
ax[0].legend()
ax[0].set_title('Região delimitada')

theta = np.linspace(0, 2 * np.pi, 100)
Y, Theta = np.meshgrid(y_vals, theta)
X = np.sqrt(Y) * np.cos(Theta)
Z = np.sqrt(Y) * np.sin(Theta)

ax[1] = plt.axes(projection='3d')
ax[1].plot_surface(X, Y, Z, color='skyblue', alpha=0.4)
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$')
ax[1].set_zlabel('$z$')
ax[1].set_title('Sólido de revolução')

plt.show()
print(f'O volume do sólido obtido pela rotação é: {volume_val:.4f}')
```

Para os problemas a e b, determine a média, usando a integração, para a funções nos intervalos especificados.

a. $f(x) = 8x - 3 + 5e^{2-x}$ no intervalo $[0, 2]$ - Vale 1 ponto

b. Considere que a posição de um objeto em movimento ao longo do eixo $ x $ é descrita pela função de deslocamento $ x(t) = t^3 - 6t^2 + 9t + 1 $, onde $ x(t) $ está em metros e $t$ está em segundos. Determine a velocidade média do objeto no intervalo de tempo de $ t = 1$ segundo até $t = 4$ segundos. - Vale 2 pontos

##**RESPOSTA 6**

**PARTE** (a): Média da função $$( f(x) = 8x - 3 + 5e^{2-x} )$$ no intervalo [0, 2]


Para encontrar a média de uma função \( f(x) \) no intervalo \([a, b]\), se usa a fórmula:
$$ \text{Média} = \frac{1}{b-a} \int_a^b f(x) \, dx $$

Para \( f(x) = 8x - 3 + 5e^{2-x} \) no intervalo \([0, 2]\):
$$ \text{Média} = \frac{1}{2-0} \int_0^2 (8x - 3 + 5e^{2-x}) \, dx $$
$$ \text{Média} = \frac{1}{2} \int_0^2 (8x - 3 + 5e^{2-x}) \, dx $$

Calculo da integral:

$$ \int_0^2 (8x - 3 + 5e^{2-x}) \, dx $$

$$ = \left[ 4x^2 - 3x - 5e^{2-x} \right]_0^2 $$

$$ = \left( 4(2)^2 - 3(2) - 5e^{2-2} \right) - \left( 4(0)^2 - 3(0) - 5e^{2-0} \right) $$

$$ = \left( 16 - 6 - 5 \right) - \left( -5e^2 \right) $$

$$ = 20 - 5e^2 $$

$$ \text{Média} = \frac{1}{2} \times (20 - 5e^2) $$

$$ = 20.9726 $$

Portanto, a média da função f(x) no intervalo [0, 2] é aproximadamente  20.9726 .


**PARTE** (b): Velocidade média do objeto

A posição do objeto é dada por $$( x(t) = t^3 - 6t^2 + 9t + 1 )$$
$$ \text{Velocidade média} = \frac{x(t_2) - x(t_1)}{t_2 - t_1} $$

Para o intervalo de tempo de ( t = 1 ) segundo até ( t = 4 ) segundos:
$$ \text{Velocidade média} = \frac{x(4) - x(1)}{4 - 1} $$

Clcular \( x(4) \) e \( x(1) \):

$$ x(4) = 4^3 - 6 \cdot 4^2 + 9 \cdot 4 + 1 = 64 - 96 + 36 + 1 = 5 $$

$$ x(1) = 1^3 - 6 \cdot 1^2 + 9 \cdot 1 + 1 = 1 - 6 + 9 + 1 = 5 $$

$$ \text{Velocidade média} = \frac{5 - 5}{4 - 1} = \frac{0}{3} = 0 \, \text{m/s} $$

Portanto, a velocidade média do objeto no intervalo de tempo de ( t = 1 ) segundo até ( t = 4) segundos é ( 0 m/s).


```python
import sympy as sp

# Parte (a)
x = sp.symbols('x')
f = 8*x - 3 + 5*sp.exp(2-x)
integral_f = sp.integrate(f, (x, 0, 2))
media_f = (1/2) * integral_f
media_f_eval = media_f.evalf()

print(f'Média de f(x) no intervalo [0, 2]: {media_f_eval:.4f}')

# Parte (b)
t = sp.symbols('t')
x_t = t**3 - 6*t**2 + 9*t + 1
x_4 = x_t.subs(t, 4)
x_1 = x_t.subs(t, 1)
velocidade_media = (x_4 - x_1) / (4 - 1)
velocidade_media_eval = velocidade_media.evalf()

print(f'Velocidade média do objeto no intervalo [1, 4] segundos: {velocidade_media_eval:.4f} m/s')
```

# Questão 7 - Vale até 2 pontos

Nos ítens a, b a seguir, encontre a família de antiderivadas das seguintes funções:

a. $f(x) = 4x^3 - 6x^2 + 2x - 5 $.

b. $f(x) = \frac{1}{x} + e^x$ para todo $x\neq 0$.




##**RESPOSTA 7**

**PARTE** (a): Antiderivada de $$( f(x) = 4x^3 - 6x^2 + 2x - 5 )$$

Tem que integrar cada termo da função separadamente:

$$ \int (4x^3 - 6x^2 + 2x - 5) \, dx $$

$$ \int 4x^3 \, dx = \frac{4x^4}{4} = x^4 $$

$$ \int -6x^2 \, dx = -6 \cdot \frac{x^3}{3} = -2x^3 $$

$$ \int 2x \, dx = 2 \cdot \frac{x^2}{2} = x^2 $$

$$ \int -5 \, dx = -5x $$

Somando essas antiderivadas e adicionando a constante de integração \( C \), obtemos:

$$ \int (4x^3 - 6x^2 + 2x - 5) \, dx = x^4 - 2x^3 + x^2 - 5x + C $$


**PARTE** (b): Antiderivada de $$( f(x) = \frac{1}{x} + e^x )$$ para todo $$( x \neq 0 )$$

Tem que integrar cada termo separadamente.

$$ \int \left( \frac{1}{x} + e^x \right) \, dx $$

$$ \int \frac{1}{x} \, dx = \ln|x| $$

$$ \int e^x \, dx = e^x $$

Somando essas antiderivadas e adicionando a constante de integração \( C \), obtemos:

$$ \int \left( \frac{1}{x} + e^x \right) \, dx = \ln|x| + e^x + C $$
