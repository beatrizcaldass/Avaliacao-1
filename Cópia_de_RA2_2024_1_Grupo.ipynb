##Problema 1
Entre todos os retângulos possíveis, de area igual a 100. Encontre aquele com o menor perímetro.



##Resposta 1


```python
import sympy as sp

#definindo variável simbólica
x = sp.symbols('x')

#definindo a função para o perímetro em termos de x
P = 2 * (100 / x + x)

#encontrando a derivada da função em relação a x
dP_dx = sp.diff(P, x)

#encontrando os valores críticos de x
x_critical = sp.solve(dP_dx, x)

#selecionando apenas os valores positivos de x
x_critical = [val for val in x_critical if val > 0]

#avaliando o perímetro para os valores críticos de x
P_minimum = min([P.subs(x, x_val) for x_val in x_critical])


print("Valor mínimo de x:", float(x_critical[0]))
print("Perímetro mínimo:", float(P_minimum))
```

##Problema 2
Você foi encarregado de determinar o melhor preço e o número de itens que devem ser vendidos para maximizar o lucro da sua empresa. Uma pesquisa indicou que se você fixar o preço em $1,50$, será capaz de vender 5000 itens. Contudo, a cada $0.10$ de redução no preço de venda você terá um aumento de 1000, no múmero de itens vendidos. Considere que seus custos fixos são de $2000.00$ e que existe um custo marginal de $0.50$ por item produzido. Encontre o preço de venda que gerará o maior lucro.

##Resposta 2


```python
import sympy as sp

#definindo variáveis simbólicas
p = sp.symbols('p')

#expressão para o número de itens vendidos em função do preço de venda
n = 5000 + 1000 * ((1.50 - p) / 0.10)

#expressão para o lucro
L = p * n - (2000 + 0.5 * n)

#encontrando o máximo da função de lucro
max_profit_price = sp.solve(sp.diff(L, p), p)

#avaliando o máximo da função de lucro
max_profit = L.subs(p, max_profit_price[0])

#imprimindo o resultado
print("Preço que maximiza o lucro:", float(max_profit_price[0]))
print("Lucro máximo:", float(max_profit))
```

## Problema 3
Você é responsável pelo projeto de criação de uma linha de distribuição de energia que irá conectar um gerador eólico instalado no mar a 1km da costa, com uma central de distribuição, no solo, localizada a 5km da estação eólica, como pode ser visto na figura a seguir: eolica.jpg

Considere que terá um custo de $50.00$ a cada metro de instalação na terra e $130.00$ a cada metro de instalação da linha de transmissão sob o oceano. Determine o comprimento que deve ser percorrido em terra para minimizar o custo total de instalação.

##Resposta 3

```python
import sympy as sp

#definindo a variável simbólica
x = sp.symbols('x')

#definindo a função de custo total
C = 50 * x + 130 * (1000 - x)

#calculando a derivada da função de custo
C_prime = sp.diff(C, x)

#encontrando a solução para a derivada
roots = sp.solve(C_prime, x)

#inicializando a variável para armazenar o comprimento ótimo
optimal_x = None

#verificando se as raízes estão dentro do intervalo de interesse (0 a 1000 metros)
for root in roots:
    if 0 <= root <= 1000:
        optimal_x = root
        break

#verificando se uma solução foi encontrada e calculando o custo total correspondente
if optimal_x is not None:
    total_cost = 50 * optimal_x + 130 * (1000 - optimal_x)
    print(f"Comprimento em terra: {optimal_x} metros")
    print(f"Custo total correspondente: ${total_cost:.2f}")
else:
    print("Não há uma solução dentro do intervalo especificado.")
```

##Problema 4
Considerando a função: $$f_{(x)} = x^4-6x^2$$


a) Trace seu gráfico, encontre suas raízes, determine se existe pelo menos um intervalo onde  existam pontos de máximo, mínimo ou sela e, caso estes pontos criticos, máximo local, mínimo local ou sela existam encontre estes pontos.

b) Caso os pontos críticos, máximo local, mínimo local ou sela existam, trace o gráfico da função, da sua primeira derivada, da sua segunda derivada e marque, neste gráfico, os pontos críticos encontrados (máximo, mínimo, sela) se e onde eles existirem.

##Resposta 4


```python
import numpy as np
import matplotlib.pyplot as plt

#definindo a função
def f(x):
    return x**4 - 6*x**2

x_values = np.linspace(-3, 3, 400)
y_values = f(x_values)


plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label='$f(x) = x^4 - 6x^2$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função $f(x) = x^4 - 6x^2$')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()

#encontrando raízes da função
roots = np.roots([1, 0, -6, 0])
real_roots = [root for root in roots if np.isreal(root)]
for root in real_roots:
    plt.plot(root, f(root), 'ro')
    plt.text(root, f(root), f'({root.real:.2f}, {f(root):.2f})', fontsize=10, ha='right')

#encontrando pontos críticos
critical_points = np.roots([4, 0, -6])
real_critical_points = [point for point in critical_points if np.isreal(point)]
for point in real_critical_points:
    if point.imag == 0:
        plt.plot(point, f(point), 'go')
        plt.text(point, f(point), f'({point.real:.2f}, {f(point):.2f})', fontsize=10, ha='right')

plt.show()
```

```python
import sympy as sp

#definindo a variável simbólica
x = sp.symbols('x')

#definindo a função
f = x**4 - 6*x**2

#calculando a primeira e segunda derivadas
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

#encontrando os pontos críticos
critical_points = sp.solve(f_prime, x)

minima = []
maxima = []
saddle = []

for point in critical_points:
    if f_double_prime.subs(x, point) > 0:
        minima.append(point)
    elif f_double_prime.subs(x, point) < 0:
        maxima.append(point)
    else:
        saddle.append(point)


print("Pontos Críticos:")
for point in critical_points:
    print(f"    x = {sp.pretty(point)}")

print("Classificação dos Pontos Críticos:")
if minima:
    print("    Mínimo(s):", [sp.pretty(minimum) for minimum in minima])
else:
    print("    Não há mínimos")
if maxima:
    print("    Máximo(s):", [sp.pretty(maximum) for maximum in maxima])
else:
    print("    Não há máximos")
if saddle:
    print("    Ponto(s) de Sela:", [sp.pretty(s) for s in saddle])
else:
    print("    Não há pontos de sela")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#definindo a função, primeira e segunda derivadas
def f(x):
    return x**4 - 6*x**2

def f_prime(x):
    return 4*x**3 - 12*x

def f_double_prime(x):
    return 12*x**2 - 12

x_values = np.linspace(-3, 3, 400)
y_values_f = f(x_values)
y_values_f_prime = f_prime(x_values)
y_values_f_double_prime = f_double_prime(x_values)

#encontrando os pontos críticos
critical_points = np.roots([4, 0, -6])

minima = []
maxima = []
saddle = []

for point in critical_points:
    if f_double_prime(point) > 0:
        minima.append(point)
    elif f_double_prime(point) < 0:
        maxima.append(point)
    else:
        saddle.append(point)

plt.figure(figsize=(12, 8))

#função original
plt.plot(x_values, y_values_f, label='Função Original: $f(x) = x^4 - 6x^2$')

#primeira derivada
plt.plot(x_values, y_values_f_prime, label='Primeira Derivada: $f\'(x) = 4x^3 - 12x$')

#segunda derivada
plt.plot(x_values, y_values_f_double_prime, label='Segunda Derivada: $f\'\'(x) = 12x^2 - 12$')

#marcando os pontos críticos
plt.plot(minima, f(np.array(minima)), 'bo', label='Mínimos')
plt.plot(maxima, f(np.array(maxima)), 'go', label='Máximos')
plt.plot(saddle, f(np.array(saddle)), 'ro', label='Pontos de Sela')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico da Função e suas Derivadas')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

##Problema 5

##Resposta 5


```python
import numpy as np
import matplotlib.pyplot as plt

#definindo a função
def f(x):
    return 3*np.sin(x) + 2*np.cos(x) + 0.5*x**3 - 2*x**2 + 3*x + 1


x_values = np.linspace(-10, 10, 400)
y_values = f(x_values)

#encontrando as raízes da função
roots = np.roots([0.5, -2, 3, 1])

#calculando a primeira e segunda derivadas
def f_prime(x):
    return 1.5*x**2 - 4*x + 3 - 3*np.sin(x) + 2*np.cos(x)

def f_double_prime(x):
    return 3*x - 4 + 2*np.sin(x) + 3*np.cos(x)

#encontrando os pontos críticos
critical_points = np.roots([1.5, -4, 3])


minima = []
maxima = []
saddle = []

for point in critical_points:
    if f_double_prime(point) > 0:
        minima.append(point)
    elif f_double_prime(point) < 0:
        maxima.append(point)
    else:
        saddle.append(point)


plt.figure(figsize=(12, 8))
plt.plot(x_values, y_values, label='f(x) = $3\sin(x) + 2\cos(x) + 0.5x^3 - 2x^2 + 3x + 1$')
plt.scatter(roots, f(roots), color='red', label='Raízes')

if minima:
    plt.scatter(minima, f(np.array(minima)), color='blue', label='Mínimos Locais')
if maxima:
    plt.scatter(maxima, f(np.array(maxima)), color='green', label='Máximos Locais')
if saddle:
    plt.scatter(saddle, f(np.array(saddle)), color='purple', label='Pontos de Sela')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da Função e suas Raízes')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```
